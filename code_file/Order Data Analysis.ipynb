### Read data from the file and handle null values

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Siddharth Singh\Desktop\Â \Python+Sql Project\orders.csv",na_values=['Not Available', 'unknown'])
df.head(10)
```

### Rename columns names, make them lower case and replace space with underscore

```python
# method 1: you have to do individually for each column name
# df.rename(columns={'Order Id':'order_id','City':'city'})
```

```python
# method 2: using str function
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df
```

### Drive new columns: discount, sale_price and profit

```python
df['discount']=df['list_price']*df['discount_percent']*0.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

### Convert order_date from object data type to datetime

```python
# Currently an object dtype
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
# now a datetime64[ns] column
```

### Drop cost price, list price and discount percent columns

```python
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
df
```
